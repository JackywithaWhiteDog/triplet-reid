```python
%matplotlib inline

from matplotlib import pyplot as plt

plt.style.use('ggplot')
```

```python
logs = [
    "mAP: 2.52% | top-1: 10.10% | top-2: 13.90% | top-5: 20.61% | top-10: 26.69%",
    "mAP: 58.78% | top-1: 79.16% | top-2: 85.45% | top-5: 90.65% | top-10: 94.00%",
    "mAP: 58.34% | top-1: 78.92% | top-2: 85.15% | top-5: 91.03% | top-10: 94.36%",
    "mAP: 60.29% | top-1: 79.60% | top-2: 85.75% | top-5: 92.01% | top-10: 95.13%",
    "mAP: 68.88% | top-1: 85.33% | top-2: 89.79% | top-5: 94.39% | top-10: 96.62%",
    "mAP: 70.11% | top-1: 85.39% | top-2: 89.99% | top-5: 94.39% | top-10: 96.59%",
]
keys = [l.split(":")[0] for l in logs[0].split(" | ")]
values = [
    {k: float(l.split(":")[-1].strip()[:-1]) for k, l in zip(keys, log.split(" | "))}
    for log in logs
]
```

```python
keys
```

```python
epochs = range(0, 25000+1, 5000)
plt.figure(dpi=100)
plt.plot(epochs[1:], [v[keys[0]] for v in values[1:]], "-o")
plt.xlabel("Epoch")
plt.ylabel("mAP")
plt.title("mAP")
plt.show()
```

```python
epochs = range(0, 25000+1, 5000)
plt.figure(dpi=100)
for k in keys[1:]:
    plt.plot(epochs[1:], [v[k] for v in values[1:]], "-o")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.title("Top-N Accuracy")
plt.legend(keys[1:])
plt.show()
```
